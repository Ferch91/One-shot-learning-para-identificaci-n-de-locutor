```python

def loadwavs(path,top,p=False):
    
    '''
    path = directorio , top = cantidad maxima de archivos por locutor , p = imprimir cantida de archivos por locutor
    Funcion que lee los archivos dentro de un directorio (path) y por locutor devuelve un numero determinado por top de
    archivos wav por cada locutor, todos los archivos se concatenan en un unico vector X. Tambien se devuelve el vector
    y que contiene el numero correspondiente al id del locutor . ejemplo: id10294 devuelve 294.
    
    '''
    
    n = 0
    X=[]
    y = []
    cat_dict = {}
    lang_dict = {}
    curr_y = n
    files = []
    j=0
    
    
        # Obtenemos directorios
    for locutor in os.listdir(path):
        print("Cargando Locutor: " + locutor)
        lang_dict[locutor] = [curr_y,None]
        locutor_path = os.path.join(path,locutor)
        i=0
            # Dado que dentro de cada locutor hay diferentes carpetas se lo carga por separado
        for carpeta in os.listdir(locutor_path):
            cat_dict[curr_y] = (locutor, carpeta)
            carpeta_path = os.path.join(locutor_path, carpeta)
                # Leemos todos los archivos en el directorio
            for filename in os.listdir(carpeta_path):
                
                if i<top:
                    
                    wav_path = os.path.join(carpeta_path, filename)
                    wav_path_1 = os.path.join(carpeta_path, filename[0:5])

                    fs, wav = wavfile.read(wav_path)
            
                    if np.shape(wav)[0]>80000:
                        X.append(wav[0:80000])
                        files.append(wav_path_1)
                        y.append(locutor[4:7])
                        i=i+1
                
            curr_y += 1
        j=j+1
        if p==True:
            
            print('Numero total de archivos:' ,i) 
            print('')

    print('')
    print('---------------------------------------------------------------')
    print('Numero total de locutores:',j)
    print('Cantidad de archivos por locutor:',i)
    print('Cantidad total de archivos:',i*j)
    print('---------------------------------------------------------------')


    print('')
    print('Hecho')
    print('')
    return X,y,j,files
```

```python
def loadubmlist(path,top,p=False):
    
    '''
    path = directorio , top = cantidad maxima de archivos por locutor , p = imprimir cantida de archivos por locutor
    Funcion que lee los archivos dentro de un directorio (path) y por locutor devuelve un numero determinado por top de
    archivos wav por cada locutor, todos los archivos se concatenan en un unico vector X. Tambien se devuelve el vector
    y que contiene el numero correspondiente al id del locutor . ejemplo: id10294 devuelve 294.
    
    '''
    
    n = 0
    X=[]
    y = []
    cat_dict = {}
    lang_dict = {}
    curr_y = n
    files = []
    j=0
    
    
        # Obtenemos directorios
    for locutor in os.listdir(path):
        #print("Cargando Locutor: " + locutor)
        lang_dict[locutor] = [curr_y,None]
        locutor_path = os.path.join(path,locutor)
        i=0
            # Dado que dentro de cada locutor hay diferentes carpetas se lo carga por separado
        for carpeta in os.listdir(locutor_path):
            cat_dict[curr_y] = (locutor, carpeta)
            carpeta_path = os.path.join(locutor_path, carpeta)
                # Leemos todos los archivos en el directorio
            for filename in os.listdir(carpeta_path):
                
                if i<top:
                    
                    wav_path = os.path.join(carpeta_path, filename)

                    files.append(wav_path)
                    i=i+1
                
            curr_y += 1
        j=j+1
        if p==True:
            
            print('Numero total de archivos:' ,i) 
            print('')

    print('')
    print('---------------------------------------------------------------')
    print('Numero total de locutores:',j)
    print('Cantidad de archivos por locutor:',i)
    print('Cantidad total de archivos:',i*j)
    print('---------------------------------------------------------------')


    print('')
    print('Hecho')
    print('')
    return X,y,j,files
```

```python
# Se define extractor para MFCC



def mfcc():
    extractor = sidekit.FeaturesExtractor(audio_filename_structure="{}.wav",
                                          feature_filename_structure="/home/fernando/anaconda3/envs/TFM Master IA/audio/IV_FULL_H5/{}.h5",
                                          sampling_frequency=16000,
                                          lower_frequency=100,
                                          higher_frequency=3700,
                                          filter_bank="log",
                                          filter_bank_size=26,
                                          window_size=0.025,
                                          shift=0.01,
                                          ceps_number=7,
                                          vad="snr",
                                          snr=40,
                                          pre_emphasis=0.97,
                                          save_param=["vad", "cep"],
                                          keep_all_features=False)
    channel_list=[0]*len(files_path)
    extractor.save_list(show_list=files_path,channel_list=channel_list,num_thread=4)    

# Se define servidor necesario para el procedimiento de MFCC e Ivectores por sidekit

    
def get_feature_server():
    
    features_server = sidekit.FeaturesServer(features_extractor=None,
                                           feature_filename_structure=None,
                                           sources=None,
                                           dataset_list=["cep", "vad"],
                                           mask=None,
                                           feat_norm="cmvn",
                                           global_cmvn=None,
                                           dct_pca=False,
                                           dct_pca_config=None,
                                           sdc=True,
                                           sdc_config=None,
                                           delta=False,
                                           double_delta=False,
                                           delta_filter=None,
                                           context=None,
                                           traps_dct_nb=None,
                                           rasta=True,
                                           keep_all_features=False)
    return features_server
    
```

```python
#MODELING THE UBM AND T MATRIX FOR I VECTORS

print("Obtaining MFCC...")
mfcc()
print("Done")
print("")

print("Audio Preprocessing H5....")
server=get_feature_server()
print("Done")
print("")


```

```python
#Se entrena el modelo para I/vectores y luego se los extrae

print("Audio Preprocessing....")
server=get_feature_server()
print("Done")
print("")

# Leemos el UBM que se entreno y guardo previamente
print("Reading UBM previously trained...")
ubm=sidekit.Mixture()
ubm.read('/home/fernando/anaconda3/envs/TFM Master IA/audio/UBM/ubm.h5')


print("Mapping wavs to models...")
             
tv_data_dir="/home/fernando/anaconda3/envs/TFM Master IA/audio/EVAL_H5/wav_train/"
tv_list=os.listdir(tv_data_dir)
tv_models=np.array([files.split("_")[0] for files in tv_list]) #Nombres span eng
tv_segments=np.array([files.split(".")[0] for files in tv_list]) #span_1 eng_1

tv_idmap=sidekit.IdMap()
tv_idmap.leftids=np.asarray(tv_models)
tv_idmap.rightids=np.asarray(tv_segments)
tv_idmap.start=np.empty(tv_idmap.rightids.shape,"|O")
tv_idmap.stop=np.empty(tv_idmap.rightids.shape,"|O")
print("")


print("Computing Statistics....")
tv_stat=sidekit.StatServer(statserver_file_name=tv_idmap,ubm=ubm)
tv_stat.accumulate_stat(ubm=ubm,feature_server=server,seg_indices=range(tv_stat.segset.shape[0]),num_thread=4)
tv_stat.write("/home/fernando/anaconda3/envs/TFM Master IA/audio/stativec1.h5")


fa = sidekit.FactorAnalyser()

fa.total_variability_single('/home/fernando/anaconda3/envs/TFM Master IA/audio/stativec1.h5',
                            ubm,
                            tv_rank=400,
                            nb_iter=8,
                            min_div=False,
                            tv_init=None,
                            batch_size=100,
                            save_init=False,
                            output_file_name='/home/fernando/anaconda3/envs/TFM Master IA/audio/TV_Matrix')


print('hecho')

```

```python
# Se extraen los Ivectores en la variable iv_train

print("Extracting Ivectors...")
iv_train = fa.extract_ivectors(ubm=ubm,stat_server_filename="/home/fernando/anaconda3/envs/TFM Master IA/audio/stativec.h5",num_thread=4)
print("Done")
```
